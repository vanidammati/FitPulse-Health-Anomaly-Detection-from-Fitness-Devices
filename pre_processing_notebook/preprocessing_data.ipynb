```python
import pandas as pd
import numpy as np

# ===============================
# 1️⃣ LOAD DATASET
# ===============================
df = pd.read_csv("Fitness_Health_Tracking_Dataset.csv")

print("Dataset Loaded Successfully!")
print("Shape:", df.shape)

# ===============================
# 2️⃣ CLEAN COLUMN NAMES
# ===============================
df.columns = df.columns.str.strip()          # remove spaces
df.columns = df.columns.str.replace(" ", "_") # replace spaces with _
df.columns = df.columns.str.lower()          # lowercase

print("\nCleaned Column Names:")
print(df.columns)

# ===============================
# 3️⃣ AUTO-DETECT DATE COLUMN
# ===============================
date_col = None

for col in df.columns:
    if "date" in col or "time" in col:
        date_col = col
        break

if date_col is None:
    raise Exception("❌ No Date/Time column found in dataset.")

print("\nDetected Date Column:", date_col)

# ===============================
# 4️⃣ CONVERT TO DATETIME & SET INDEX
# ===============================
df[date_col] = pd.to_datetime(df[date_col], errors='coerce')
df = df.dropna(subset=[date_col])  # remove invalid dates
df = df.sort_values(date_col)
df = df.set_index(date_col)

print("Date column converted and set as index.")

# ===============================
# 5️⃣ CHECK NULL VALUES BEFORE CLEANING
# ===============================
print("\nNull Values BEFORE Cleaning:")
print(df.isnull().sum())
print("Total Nulls Before:", df.isnull().sum().sum())

# ===============================
# 6️⃣ HANDLE NUMERIC COLUMNS
# ===============================
numeric_cols = df.select_dtypes(include=np.number).columns

# Linear Interpolation
df[numeric_cols] = df[numeric_cols].interpolate(method='linear')

# Forward Fill
df[numeric_cols] = df[numeric_cols].ffill()

# Backward Fill
df[numeric_cols] = df[numeric_cols].bfill()

# ===============================
# 7️⃣ HANDLE CATEGORICAL COLUMNS
# ===============================
categorical_cols = df.select_dtypes(exclude=np.number).columns

df[categorical_cols] = df[categorical_cols].ffill()
df[categorical_cols] = df[categorical_cols].bfill()

# ===============================
# 8️⃣ FINAL NULL CHECK
# ===============================
print("\nNull Values AFTER Cleaning:")
print(df.isnull().sum())
print("Total Nulls After:", df.isnull().sum().sum())

# ===============================
# 9️⃣ RESET INDEX (Optional)
# ===============================
df = df.reset_index()

# ===============================
# 🔟 SAVE CLEANED DATASET
# ===============================
df.to_csv("cleaned_fitness_data.csv", index=False)

print("\n✅ Data cleaned successfully!")
print("✅ Cleaned file saved as: cleaned_fitness_data.csv")
```
